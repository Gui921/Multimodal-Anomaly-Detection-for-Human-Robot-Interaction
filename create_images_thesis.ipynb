```python
import matplotlib.pyplot as plt

# Data
labels = [
    'Move no cup',
    'Human Handover',
    'Move with cup',
    'place',
    'Robot picking object',
    'Robot Handover',
    'Idle'
]

sizes = [710, 222, 593, 317, 340, 209, 193]

# Custom function to display both number and percentage
def autopct_func(pct, allvals):
    absolute = int(round(pct/100.*sum(allvals)))
    return f"{absolute} ({pct:.1f}%)"

# Plot
plt.figure(figsize=(8, 8))
plt.pie(
    sizes,
    labels=labels,
    autopct=lambda pct: autopct_func(pct, sizes),
    startangle=140,
    textprops={'fontsize': 12}
)
plt.title('Activity Distribution', fontsize=18, pad=40)
plt.axis('equal')

plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

vision_only_data = np.load('roc_vision.npz')
fpr_vision_only = vision_only_data["fpr"]
tpr_vision_only = vision_only_data["tpr"]
roc_auc_vision_only = vision_only_data["roc_auc"].item()

vision_sgg = np.load('roc_video_sgg.npz')
fpr_vision_sgg = vision_sgg["fpr"]
tpr_vision_sgg = vision_sgg["tpr"]
roc_vision_sgg = vision_sgg["roc_auc"].item()

vision_and_bag = np.load('roc_vision_with_bag.npz')
fpr_vision_and_bag = vision_and_bag["fpr"]
tpr_vision_and_bag = vision_and_bag["tpr"]
roc_auc_vision_and_bag = vision_and_bag["roc_auc"].item()

all = np.load('roc_all.npz')
fpr_all = all['fpr']
tpr_all = all['tpr']
roc_auc_all = all["roc_auc"]

baseline = np.load('roc_baseline.npz')
fpr_baseline = baseline['fpr']
tpr_baseline = baseline['tpr']
roc_auc_baseline = baseline['roc_auc']
```

```python
plt.figure(figsize=(7, 7))

plt.plot(fpr_baseline, tpr_baseline,
         label=f"Baseline (AUC = {roc_auc_baseline:.2f})")

# Vision only
plt.plot(fpr_vision_only, tpr_vision_only,
         label=f"Vision Only (AUC = {roc_auc_vision_only:.2f})")

# Vision SGG
plt.plot(fpr_vision_sgg, tpr_vision_sgg,
         label=f"Vision + SGG (AUC = {roc_vision_sgg:.2f})")

# Vision + Bag
plt.plot(fpr_vision_and_bag, tpr_vision_and_bag,
         label=f"Vision + Sensor (AUC = {roc_auc_vision_and_bag:.2f})")

# All features
plt.plot(fpr_all, tpr_all,
         label=f"All Features (AUC = {roc_auc_all:.2f})")

# Random baseline
plt.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Random")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves Comparison")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
